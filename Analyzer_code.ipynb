```python
import os
import cvlib as cv
import numpy as np
import cv2
from IPython.display import display, Image

image_paths = [
    r"D:\SEM V\AINN\assignment\images\girltwo.jpg",
    r"D:\SEM V\AINN\assignment\images\eleven.jpg",
    r"D:\SEM V\AINN\assignment\images\man.jpg",
    r"D:\SEM V\AINN\assignment\images\group.jpg",
    r"D:\SEM V\AINN\assignment\images\mantwo.jpg"
]

# Function to process each image and perform face & gender detection
def process_image(image_path):
    if not os.path.exists(image_path):
        print(f"[ERROR] The file at {image_path} does not exist.")
        return
    
    image = cv2.imread(image_path)
    if image is None:
        print(f"[ERROR] Unable to read the image file at {image_path}.")
        return
    
    print(f"Processing image: {image_path}")
    print(f"Input Image Shape: {image.shape}")

    # Apply Gaussian Blur to reduce noise
    blurred_image = cv2.GaussianBlur(image, (5, 5), 0)

    # Apply face detection
    faces, confidences = cv.detect_face(blurred_image)
    print(f"Detected Faces: {faces}")
    print(f"Confidences: {confidences}")

    padding = 10  # Add padding around the detected faces

    # Loop through detected faces
    for face, conf in zip(faces, confidences):
        if conf > 0.5:  # Confidence threshold for face detection
            (startX, startY) = max(0, face[0] - padding), max(0, face[1] - padding)
            (endX, endY) = min(image.shape[1] - 1, face[2] + padding), min(image.shape[0] - 1, face[3] + padding)

            # Draw rectangle over face
            cv2.rectangle(image, (startX, startY), (endX, endY), (0, 255, 0), 2)
            face_crop = np.copy(blurred_image[startY:endY, startX:endX])

            # Resize the cropped face to 100x100
            face_crop = cv2.resize(face_crop, (100, 100))

            # Apply gender detection using the cvlib model
            labels, confs = cv.detect_gender(face_crop)
            print(f"Gender Labels: {labels}")
            print(f"Gender Confidences: {confs}")

            # Get the index of the label with maximum confidence score
            idx = np.argmax(confs)  # Correct: Use np.argmax for highest confidence
            label = labels[idx]  # Select the label with highest confidence

            # Write label name (male/female) on top of the bounding box
            cv2.putText(image, label, (startX, startY - 10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # Save and display the output image
    output_image_path = f"gender_detection_output_{os.path.basename(image_path)}"
    cv2.imwrite(output_image_path, image)
    print(f"Output image saved to {output_image_path}")

    # Display the output image in Jupyter Notebook
    display(Image(output_image_path))

# Loop through the list of image paths and process each one
for img_path in image_paths:
    process_image(img_path)
```

```python


import cv2
import cvlib as cv
import numpy as np
import matplotlib.pyplot as plt

# Initialize the Haar Cascade for face detection (pre-trained model)
face_detector = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

def find_faces(image_path):
    
    image = cv2.imread(image_path)

    if image is None:
        print(f"Error: Could not load image at {image_path}.")
        return None, None

    # Convert the image to grayscale (face detection works better on grayscale images)
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    large_faces = face_detector.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=6, minSize=(120, 120))

    small_faces = face_detector.detectMultiScale(gray_image, scaleFactor=1.04, minNeighbors=4, minSize=(40, 40))

    # Combine both large and small face detections
    detected_faces = list(large_faces) + list(small_faces)

    return detected_faces, image

# Function to predict gender of detected faces
def detect_gender(image, faces):
    for (x, y, w, h) in faces:
        # Extract the face from the image
        face_crop = np.copy(image[y:y+h, x:x+w])

        # Use cvlib's detect_gender function to predict gender
        label, confidence = cv.detect_gender(face_crop)

        # Get the label with the highest confidence
        idx = np.argmax(confidence)
        gender = label[idx]
        gender_confidence = confidence[idx]

        # Draw a rectangle around the face
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 3)

        # Put the predicted gender with confidence on the image
        cv2.putText(image, f"{gender} ({gender_confidence*100:.2f}%)", (x, y - 10), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 0), 2)

image_path = r"D:\SEM V\AINN\assignment\images\id.jpg"  
faces, image = find_faces(image_path)

if faces is not None and len(faces) > 0:
    # Predict and display gender for each detected face
    detect_gender(image, faces)

    # Convert the image from BGR (OpenCV format) to RGB (Matplotlib format) for display
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Display the image using Matplotlib
    plt.figure(figsize=(8, 6))
    plt.imshow(image_rgb)
    plt.axis('off')  
    plt.show()
else:
    print("No faces were found.")
```

```python

import cv2
import matplotlib.pyplot as plt

# Load the pre-trained Haar Cascade classifier for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

def detect_faces(image_path):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    return faces, img 

def display_image_with_faces(image_path):
    faces, img = detect_faces(image_path)

    if faces is not None and len(faces) > 0:
        # Draw rectangles around the faces
        for (x, y, w, h) in faces:
            cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)

        # Convert BGR image to RGB for displaying in Jupyter Notebook
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        # Use matplotlib to display the image with detected faces
        plt.figure(figsize=(10, 6))
        plt.imshow(img_rgb)
        plt.axis('off')  # Hide the axes
        plt.show()
    else:
        print("No faces detected.")

# Example usage:
image_path = r"D:\SEM V\AINN\assignment\images\id.jpg"  # Replace with the path to your image
display_image_with_faces(image_path)
```
